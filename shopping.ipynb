```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
  
```

```python
shop['Age_Category']=pd.cut(shop['Age'], bins=[0,15,18,30,50,70],labels=['child','teen','young Adults','Middle-Aged Adults','old'])
shop.head()
```

```python
plt.hist(shop["Age"],bins=20)
plt.show()
```

```python
shop["Gender"].value_counts().plot(kind='bar')
plt.show()
```

```python
shop["Category"].value_counts().plot(kind='bar')
plt.show()
```

```python
shop["Item Purchased"].value_counts().plot(kind='bar')
plt.show()
```

```python
shop["Category"].value_counts().plot(kind='pie')
plt.xlabel("Category")
plt.show()
```

```python
shop.groupby('Payment Method')['Purchase Amount (USD)'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x="Payment Method",y='Purchase Amount (USD)',data=shop)
plt.show()
```

```python
shop.info()
```

```python
num_fea=shop[["Age","Purchase Amount (USD)","Review Rating","Previous Purchases"]]
num_fea.head()
```

```python
cor_mat = num_fea.corr()
cor_mat  # This will display the correlation matrix
```

```python
sns.heatmap(cor_mat)
sns.heatmap(cor_mat,annot=True,cmap="BuPu")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot age distribution
plt.hist(shop["Age"], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Shoppers')
plt.show()

# Plot gender distribution
shop["Gender"].value_counts().plot(kind='bar', color=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Shopping Trends by Gender')
plt.show()
```
